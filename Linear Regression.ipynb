## 简单线性回归问题分析

原理：仍然是让损失函数的值降到最小。首先几个数学概念：

残差/训练误差：由模型预测出的值和在训练数据集中观测到的值之间的差值为残差

预测误差/测试误差：在测试数据中预测值和观测值之间的差值叫做预测误差。

在通过训练数据集拟合线性回归曲线之后，用最小化残差的和来对拟合效果进行衡量。这种衡量模型拟合的方法叫做残差平方和（RSS）损失函数。其基本过程就是通过对所有数据集中的数据残差的平法求和来判断模型的拟合效果的。

在线性拟合曲线的时候，我们通过求损失函数的最小值来求解其中的参数。以线性回归为例，拟合曲线：y = ax + b。首先求得 “a” 的值。利用公式 a = cov(x, y) / var(x)来求解。
